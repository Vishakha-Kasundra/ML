# Data_Advertising.csv

# 1)Tv v/s Sales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

#load the data
data=pd.read_csv("Data_Advertising.csv")
# print(data)

#split the dataset
x=np.array(data.TV).reshape(-1,1)
y=np.array(data.sales).reshape(-1,1)

#load the model
linear_rmodel=linear_model.LinearRegression()

#train the model : fit
linear_rmodel.fit(x,y)

#test the model
# xtest=np.array([1.1,5.2,10.0]).reshape(-1,1)
xtest=x
ytest_result=linear_rmodel.predict(xtest)
# print(xtest)
# print(y)
# print(ytest_result)

#visualize the actual value v/s predicted value
plt.scatter(x,y,color="red", label="Actual value")
plt.plot(x,ytest_result,color="blue", label="Pridicted value")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Predicted value of TV")
plt.legend()
plt.show()
```

# 1)minTemp v/s snowfall

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn import linear_model

#load the data
data=pd.read_csv("Weather.csv")
ndf=data[['Snowfall']]
ndf.dropna(axis=0,inplace=True)
# ndf.isnull().sum()
# print(data)

#split the dataset
x=np.array(data.MinTemp).reshape(-1,1)
y=np.array(ndf).reshape(-1,1)


#load the model
linear_rmodel=linear_model.LinearRegression()

#train the model : fit
linear_rmodel.fit(x,y)

#test the model
# xtest=np.array([1.1,5.2,10.0]).reshape(-1,1)
xtest=x
ytest_result=linear_rmodel.predict(xtest)
# print(xtest)
# print(y)
# print(ytest_result)

#visualize the actual value v/s predicted value
plt.scatter(x,y,color="red", label="Actual value")
plt.plot(x,ytest_result,color="blue", label="Pridicted value")
plt.xlabel("MinTemp")
plt.ylabel("Snowfall")
plt.legend()
plt.show()
```
